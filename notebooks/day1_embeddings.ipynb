```python
def score_answer(user_answer, ideal_answer, model):
    """
    Returns semantic similarity score between user answer and ideal answer
    """
    user_emb = model.encode(user_answer)
    ideal_emb = model.encode(ideal_answer)

    score = cosine_similarity(
        [user_emb], [ideal_emb]
    )[0][0]

    return round(score, 2)
```

```python
user_answers = {
    0: "I had a very short deadline, so I planned my work carefully and delivered on time.",
    2: "I failed once by underestimating testing effort, but I learned to plan better.",
    7: "A system issue reduced model accuracy, so I investigated data drift and fixed it."
}
```

```python
results = []

for idx, qa in enumerate(questions):
    if idx in user_answers:
        score = score_answer(
            user_answers[idx],
            qa["ideal_answer"],
            model
        )

        results.append({
            "question": qa["question"],
            "user_answer": user_answers[idx],
            "ideal_answer": qa["ideal_answer"],
            "score": score
        })
```

```python
for r in results:
    print("Question:", r["question"])
    print("Score:", r["score"])
    print("-" * 60)
```
